```python
import pandas as pd
import numpy as np
import os


pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
plot.style.use('ggplot')
```

```python
# Get the current working directory
current_directory = os.getcwd()

# Specify the file name
file_name = 'LISTADO AFILIADOS.xls'

# Combine the current directory and file name
file_path = os.path.join(f"{current_directory}", file_name)
# Read the Excel file into a DataFrame
df = pd.read_excel(file_path)
# Replace 'output_file.csv' with the desired CSV file name
output_file = 'output_file.csv'

# Convert the DataFrame to CSV
df.to_csv(output_file, index=False)
```

```python
df_csv=pd.read_csv("output_file.csv")
print(df_csv.shape)
df_csv
```

```python
# Assuming df is your DataFrame
import re

# Assuming df_csv is your DataFrame
columns_list = df_csv.columns

# Use a list comprehension to replace spaces and hyphens with underscores in each column name
#The extra regex is just for the previous Ciudad_-_Departamento  correction
# Use a list comprehension with a single regular expression for all substitutions
columns_with_underscores = [re.sub(r'\s+|-|Vlr -', lambda x: '_' if x.group() == ' ' else 'Valor_', column) for column in columns_list]

# Assign the new column names to the DataFrame
df_csv.columns = columns_with_underscores

# Display the DataFrame with updated column names
df_csv
```
