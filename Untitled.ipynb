```python
# df_updated.drop(columns=["Modo_Recaudo", "Tipo_Periodo", "Razon_Social" ])
df_updated[['Observaciones', 'CodCIIU']] = df_updated[['Observaciones', 'CodCIIU']].astype(str)
df_updated.columns=df.columns.str.lower()
df_updated.to_csv("static/new_output_file.csv", index=False)
df_updated.columns
```

```python
# Columnas para la tabla empleados
empleados_columns = ['tipo_identificacion', 'identificacion', 'primer_nombre', 'segundo_nombre', 'primer_apellido', 
                     'segundo_apellido', 'tipo_sexo', 'cotizante_extranjero', 'fecha_nacimiento', 'edad', 'fecha_expedicion',
                     'incapacidades']

# Columnas para la tabla detalles_contacto
detalles_contacto_columns = ['numero_telefono', 'numero_celular', 'direccion', 
                             'ciudad_departamento', 'email', 'observaciones']

# Columnas para la tabla detalles_salario
detalles_salario_columns = ['salario_o_ibc', 'mensajeria', 'vlr_mensajeria', 'vlr_administracion',
                            'vlr_upc', 'estado', 'valor_pila', 'valor_total']

# Columnas para la tabla detalles_afiliacion
detalles_afiliacion_columns = ['fecha_afiliacion', 'identificacion_asesor', 'nombre_asesor', 
                               'valor_comision', 'afp', 'afptarifa', 'eps', 'epstarifa', 'arl', 'arltarifa', 'arl_nivel', 
                               'codciiu', 'planilla_n', 'ccf', 'ciudad_ccf', 'ccftarifa',]
```

```python
# Crear sub-DataFrames
df_empleados = df_updated[empleados_columns]
df_detalles_contacto = df_updated[detalles_contacto_columns]
df_detalles_salario = df_updated[detalles_salario_columns]
df_detalles_afiliacion = df_updated[detalles_afiliacion_columns]
df_detalles_afiliacion
```

```python
# Convert "Ninguno" to False, and everything else to True
df_empleados["cotizante_extranjero"] = df_empleados["cotizante_extranjero"].replace({"Ninguno": False}).astype(bool)

df_empleados.to_csv("static/employees.csv", index=False)
df_empleados.head()
```

```python
# Obtener la longitud del DataFrame df_detalles_contacto
num_filas = len(df_detalles_contacto)

# Generar una secuencia de números para empleado_id
# Comenzando desde 1 hasta el número de filas en el DataFrame
empleado_ids = range(1, num_filas + 1)

# Insertar la columna 'empleado_id' al principio del DataFrame
df_detalles_contacto.insert(0, 'empleado_id', empleado_ids)

# Guardar el DataFrame en un archivo CSV
df_detalles_contacto.to_csv("static/contact_details.csv", index=False)
# df_detalles_contacto.to_csv("static/contact_details.csv", index=False)
df_detalles_contacto.head()
```
